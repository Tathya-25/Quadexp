```python
from sklearn.model_selection import train_test_split

df_train,df_test=train_test_split(df,test_size=0.2,random_state=42)
```

```python
print(df_train.shape)
print(df_test.shape)
print(df.shape)
```

```python
sns.countplot(x='y',data=df_train)
plt.show()
```

```python
from sklearn.utils import resample,shuffle
df_yes=df_train[df_train["y"]=="yes"]
df_no=df_train[df_train["y"]=="no"]
df_yes_upsampled=resample(df_yes,random_state=42,n_samples=20000,replace=True)
train_new=pd.concat([df_yes_upsampled,df_no])
sns.countplot(x='y',data=train_new)
plt.show()
```

```python
train_new=train_new.sample(frac=1).reset_index(drop=True)
test_new=df_test.copy()
```

```python
train_ohe=pd.get_dummies(train_new,columns=['job','marital','education','default','housing','loan','contact','month','day_of_week','poutcome','y'])
test_ohe=pd.get_dummies(test_new,columns=['job','marital','education','default','housing','loan','contact','month','day_of_week','poutcome','y'])
print(train_ohe.shape)
print(test_ohe.shape)
```

```python
train_ohe.head()
```

```python
test_ohe.head()
```

```python
test_ohe.drop('y_yes',axis=1,inplace=True)
train_ohe.drop('y_yes',axis=1,inplace=True)
```

```python
X_train=train_ohe.drop('y_no',axis=1)
y_train=train_ohe['y_no']
X_test=test_ohe.drop('y_no',axis=1)
y_test=test_ohe['y_no']
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred_log_reg = log_reg.predict(X_test)

print("Accuracy: %.3f" %accuracy_score(y_test, y_pred_log_reg))
```

```python
from sklearn.metrics import confusion_matrix
cfm_log_reg=confusion_matrix(y_pred_log_reg,y_test)
sns.heatmap(cfm_log_reg,annot=True)
```

```python
confusion_matrix(y_pred_log_reg,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dt_gini=DecisionTreeClassifier(criterion="gini",random_state=42,max_depth=3,min_samples_leaf=5)
dt_gini.fit(X_train,y_train)
y_pred_dt_gini=dt_gini.predict(X_test)

dt_entropy=DecisionTreeClassifier(criterion="entropy",random_state=42,max_depth=3,min_samples_leaf=5)
dt_entropy.fit(X_train,y_train)
y_pred_dt_entropy=dt_entropy.predict(X_test)

print("Accuracy using gini: %.3f" %accuracy_score(y_test, y_pred_dt_gini))
print("Accuracy using entropy: %.3f" %accuracy_score(y_test, y_pred_dt_entropy))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rfc=RandomForestClassifier(n_estimators=100,bootstrap=True,random_state=42)
rfc.fit(X_train,y_train)
y_pred_rfc=rfc.predict(X_test)

print("Accuracy: %.3f" %accuracy_score(y_test, y_pred_rfc))
```

```python
from sklearn.metrics import confusion_matrix
cfm_rfc=confusion_matrix(y_pred_rfc,y_test)
sns.heatmap(cfm_rfc,annot=True)
```

```python
confusion_matrix(y_pred_rfc,y_test)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(rfc,X_train,y_train,cv=5)
print("Accuracy: %.3f" %scores.mean())
```
